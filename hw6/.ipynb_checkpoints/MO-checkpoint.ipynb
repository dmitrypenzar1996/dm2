<i> 1.1 Реализуйте методы GD, Momentum, NAG, Adagrad, Adadelta, Adam.</i>

```python

TestMOResult = namedtuple("TestMOResult", ['name', 'theta', 'func_values', 'iter_num', 'is_converged'])
```

```python
def gd(theta, dgrad, nu = 1e-2):
    new_theta = theta - nu * dgrad(theta) 
    return new_theta

def momentum(theta, dgrad, v = 0, nu = 1e-4, gamma = 0.9):
    new_v = gamma * v + nu * dgrad(theta)
    new_theta = theta - new_v
    return new_theta, new_v

def nag(theta, dgrad, v = 0, nu = 1e-4, gamma = 0.9): # nesterov accelerated gradient
    new_v = gamma * v + nu * dgrad(theta - gamma * v)
    new_theta = theta - new_v
    return new_theta, new_v

def adagrad(theta, dgrad, G = 0, nu = 0.01, eps = 1e-8):
    gt = dgrad(theta)
    G += np.square(gt)
    new_theta = theta - theta * (nu / np.sqrt(G + eps)) * gt
    return new_theta, G
    

def adadelta(theta, dgrad, EG = 0, Edtheta = 0, gamma = 0.9, eps = 10-8):
    gt = dgrad(theta)
    new_EG = gamma * EG + (1 - gamma) * np.square(gt)
    new_theta = theta - ( (Edtheta + eps) / (new_EG + eps) ) * gt 
    new_Edtheta = gamma * Edtheta + (1 - gamma) * np.square(new_theta - theta)
    
    return new_theta, EG, Etheta

def adam(theta, dgrad,  m = 0, v = 0, beta1 = 0.9, beta2 = 0.999,
        eps = 1e-8, nu = 1e-4):
    m_new = beta1 * m + (1 - beta1) * dgrad(theta)
    v_new = beta2 * v + (1 - beta2) * (np.square(dgrad(theta)))
    m_hat = m_new / (1 - beta1)
    v_hat = v_new / (1 - beta2)
    new_theta = theta - (nu / (np.sqrt(v_hat + eps))) * m_hat
    return new_theta, m_new, v_new
    
MAX_ITER = 1e6 
```

```python
def init_basic_params(f,  init_theta = None):
    if init_theta is None:
        init_theta = np.random.random(size=2)
    new_theta = init_theta
    theta = new_theta * 10000
    iter_num = 0
    new_value = f(new_theta)
    value = new_value * 10000
    return value, new_value, theta, new_theta, iter_num
```

```python
def test_gd(f, grad, init_theta = None):
    value, new_value, theta, new_theta, iter_num = init_basic_params(f, init_theta)
    values = list()
    theta_arr = list()
    while not np.allclose(value, new_value) and iter_num < MAX_ITER:
        value = new_value
        theta = new_theta
        values.append(value)
        theta_arr.append(theta)
        iter_num += 1
        new_theta = gd(theta, grad)
        new_value = f(new_theta)
    is_converged = np.allclose(value, new_value)
    return TestMOResult("gd", np.array(theta_arr), np.array(values), iter_num, is_converged)

def test_momentum(f, grad,  init_theta = None):
    value, new_value, theta, new_theta, iter_num = init_basic_params(f, init_theta)
    v = np.zeros(2)
    values = list()
    theta_arr = list()
    while not np.allclose(value, new_value) and iter_num < MAX_ITER:
        value = new_value
        theta = new_theta
        values.append(value)
        theta_arr.append(theta)
        iter_num += 1
        new_theta, v = momentum(theta, grad, v)
        new_value = f(new_theta)

    is_converged = np.allclose(value, new_value)
    return TestMOResult("momentum", np.array(theta_arr), np.array(values), iter_num, is_converged)

def test_nag(f, grad, init_theta = None):
    value, new_value, theta, new_theta, iter_num = init_basic_params(f, init_theta)
    v = np.zeros(2)
    values = list()
    theta_arr = list()
    while not np.allclose(value, new_value) and iter_num < MAX_ITER:
        value = new_value
        theta = new_theta
        values.append(value)
        theta_arr.append(theta)
        iter_num += 1
        new_theta, v = nag(theta, grad, v)
        new_value = f(new_theta)
        
    is_converged = np.allclose(value, new_value)
    
    return TestMOResult("nag", np.array(theta_arr), np.array(values), iter_num, is_converged)

def test_adagrad(f, grad,  init_theta = None):
    value, new_value, theta, new_theta, iter_num = init_basic_params(f, init_theta)
    G = np.zeros(2)
    values = list()
    theta_arr = list()
    while not np.allclose(value, new_value) and iter_num < MAX_ITER:
        value = new_value
        theta = new_theta
        values.append(value)
        theta_arr.append(theta)
        iter_num += 1
        new_theta, G = adagrad(theta, grad, G)
        new_value = f(new_theta)
    is_converged = np.allclose(value, new_value)
    return TestMOResult("adagrad", np.array(theta_arr), np.array(values), iter_num, is_converged)


def test_adadelta(f, grad,  init_theta = None):
    
    value, new_value, theta, new_theta, iter_num = init_basic_params(f, init_theta)
    EG = np.zeros(2)
    Etheta = np.zeros(2)
    values = list()
    theta_arr = list()
    while not np.allclose(value, new_value) and iter_num < MAX_ITER:
        value = new_value
        theta = new_theta
        values.append(value)
        theta_arr.append(theta)
        iter_num += 1
        new_theta, EG, Etheta = adadelta(theta, grad, EG, Etheta)
        new_value = f(new_theta)
    is_converged = np.allclose(value, new_value)
    return TestMOResult("adadelta", np.array(theta_arr), np.array(values), iter_num, is_converged)

def test_adam(f, grad,  init_theta = None):
    value, new_value, theta, new_theta, iter_num = init_basic_params(f, init_theta)
    m = np.zeros(2)
    v = np.zeros(2)
    values = list()
    theta_arr = list()

    while not np.allclose(value, new_value) and iter_num < MAX_ITER:
        value = new_value
        theta = new_theta
        values.append(value)
        theta_arr.append(theta)
        iter_num += 1
        new_theta, m, v = adam(theta, grad, m, v)
        new_value = f(new_theta)
    is_converged = np.allclose(value, new_value)
    return TestMOResult("adam", np.array(theta_arr), np.array(values), iter_num, is_converged)



def draw_mo_result(result_arr, J, func_name):
    delta = 0.025
    
    mnx = min(result.theta[:, 0].min() for result in result_arr)
    mxx = max(result.theta[:, 0].max() for result in result_arr)
    mny = min(result.theta[:, 1].min() for result in result_arr)
    mxy = max(result.theta[:, 1].max() for result in result_arr)
    
    r = delta * 50
    x = np.arange(mnx - r, mxx + r, delta)
    y =  np.arange(mny - r, mxy + r, delta)
    X, Y = np.meshgrid(x, y)
    Jvalues = np.zeros_like(X)
    for i in np.ndindex(X.shape):
        Jvalues[i] = J(np.array([X[i], Y[i]]))
    
    
    plt.figure(figsize = (7,7))
    levels = np.arange(-5, 5, 0.2)
    CS = plt.contour(X, Y, Jvalues, locator = ticker.LinearLocator(100), fontsize=1)
    
    plt.title('Result for function {}'.format(func_name))
    plt.clabel(CS, []) # no labels
    CB = plt.colorbar(CS, shrink=0.8, extend='both')
    for result in result_arr:
        plt.plot(result.theta[:, 0], result.theta[:, 1], label = result.name, alpha = 0.5, lw = 3)
        
    plt.legend(loc = "best")
    plt.show()
```

###  Часть 2: Обучение нейронной сети

```python
mnist = fetch_mldata('MNIST original')
images = mnist['data']
images = images.reshape(images.shape[0], 1, 28, 28)
labels = mnist['target'].astype(int)
np.random.seed(777)
images_train, images_test, labels_train, labels_test = train_test_split(images, labels)




for i in xrange(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(images_train[i, 0, :, :], cmap=plt.cm.Greys_r)
    plt.axis('off')
    
images_train = images_train.reshape(images_train.shape[0], 28 * 28)
images_test= images_test.reshape(images_test.shape[0], 28 * 28)
```

```python
#was copied from previous homework
def print_work_time(func):
    def wrapper(*args, **kwargs):
        start = timer()
        result = func(*args, **kwargs)
        end = timer()
        print("Execution take {}".format(end - start))
        return result
    return wrapper

@print_work_time
def learn_on_model(X, Y, model, loss_fn, optimizer, X_test = None, Y_test = None, learning_rate = 1e-2, 
                   batch_size = 100, max_epoch_num = 1, min_loss = 1e-5):    
    loss = 1000
    epoch_num = 0
     
    loss_arr = list()

    test_loss_arr = list() if not (type(X_test) is None) else None
    tot_err = 10**6
    X = X.copy()
    Y = Y.copy()

    while epoch_num < max_epoch_num:
        model = model.eval()
        #perm = np.random.permutation(X.shape[0])
        #X = X[perm]
        #Y = Y[perm]
        tot_err = 0
        sys.stdout.write("\nEpoch {} ... \n".format(epoch_num + 1))
        sys.stdout.write("Train\n")
        for bn, start in enumerate(xrange(0, X.shape[0], batch_size)):
            optimizer.zero_grad()
            if bn % 100 == 0:
                sys.stdout.write("\rBatch {}".format(bn))
            X_CUR = Variable(torch.from_numpy(X[start:start + batch_size]).type(torch.FloatTensor),
                            requires_grad = True)
            Y_CUR = Variable(torch.from_numpy(Y[start:start + batch_size]), requires_grad=False)
            output =  model.forward(X_CUR)
            loss = loss_fn(output, Y_CUR)
            loss_arr.append(loss.data.tolist()[0])
            tot_err += loss.data.tolist()[0]
            loss.backward()
            optimizer.step()
            del loss
            del X_CUR
            del Y_CUR
            del output
        sys.stdout.write("\nTest\n")
        if not (type(X_test) is None):
            model = model.train()
            for bn, start in enumerate(xrange(0, X_test.shape[0], batch_size)):
                if bn % 100 == 0:
                    sys.stdout.write("\rBatch {}".format(bn))
                X_CUR = Variable(torch.from_numpy(X_test[start:start + batch_size]).type(torch.FloatTensor),
                                requires_grad = True)
                Y_CUR = Variable(torch.from_numpy(Y_test[start:start + batch_size]), requires_grad = False)
                y_pred = model.forward(X_CUR)
                loss = loss_fn(y_pred, Y_CUR)
                test_loss_arr.append(loss.data.tolist()[0])
                del X_CUR
                del Y_CUR
                del y_pred
                del loss
        #sys.stdout.write("\n")   
        epoch_num += 1
        gc.collect() # only this've helped to save process from beeing kill
    model = model.train()
    return np.array(loss_arr), np.array(test_loss_arr)

def draw_plots(model, loss_arr, layer_arr, test_loss_arr, show_weights = False):
    x = np.arange(loss_arr.shape[0]);
    number_of_plots = 1 + sum(len(x) for x in layer_arr)
    gridSize = number_of_plots
    plt.figure(figsize=(15, 6 * number_of_plots))
    
    plotId = 1
    subpl = plt.subplot(gridSize, 1, plotId)
    if not type(test_loss_arr) is None:
        subpl.plot(x, loss_arr, x, test_loss_arr)
        subpl.legend(["Train", "Test"])
    else:
        subpl.plot(x, loss_arr)
    subpl.set_xticks(range(x.shape[0]))
    subpl.set_title("Loss function")
    subpl.set_xlabel("Epoch num")
    subpl.set_ylabel("Loss func")
    if show_weights:
        for ind1, child in enumerate(model.children()):
            for ind2, key in enumerate(child.state_dict().keys()):
                plotId += 1
                subpl = plt.subplot(gridSize, 1, plotId)
                y = layer_arr[ind1][ind2]
                subpl.plot(x, y)

                subpl.set_xticks(range(x.shape[0]))
                subpl.set_title("Layer {} : {} gradient norm".format(ind1 + 1, key))
                subpl.set_xlabel("Epoch num")
                subpl.set_ylabel(key + "_grad")
    
    plt.show()

from sklearn.metrics import classification_report

def make_classification_report(model, X_TRAIN, Y_TRAIN, X_TEST, Y_TEST, batch_size = 100):
    model = model.eval()
    print ("Train report")
    y_predicted = np.zeros(shape = (Y_TRAIN.shape[0], 10))
    
    for bn, start in enumerate(xrange(0, X_TRAIN.shape[0], batch_size)):
        if bn % 100 == 0:
            sys.stdout.write("Batch {}\n".format(bn))
        X_CUR = Variable(torch.from_numpy(X_TRAIN[start:start + batch_size]).type(torch.FloatTensor) )
        
        
        y_predicted[start:start + batch_size] = np.array(model.forward(X_CUR).data.tolist())
    indices = np.argmax(y_predicted, axis = 1)
    print (classification_report(np.array(Y_TRAIN.tolist()), indices))
    print ("Accuracy", np.mean(np.array(Y_TRAIN.tolist()) == indices))
    
    print ("Test report")
    y_predicted = np.zeros(shape = (Y_TEST.shape[0], 10))
    for bn, start in enumerate(xrange(0, X_TEST.shape[0], batch_size)):
        if bn % 100 == 0:
            sys.stdout.write("Batch {}\n".format(bn))
        X_CUR = Variable(torch.from_numpy(X_TEST[start:start + batch_size]).type(torch.FloatTensor))
        y_predicted[start:start + batch_size] = np.array(model.forward(X_CUR).data.tolist())
        del X_CUR
        
    
    
    indices = np.argmax(y_predicted, axis = 1)
    print (classification_report(np.array(Y_TEST.tolist()), indices))
    print ("Accuracy", np.mean(np.array(Y_TEST.tolist()) == indices))

    
```

```python
def FullConnectedNetwork():
    return nn.Sequential(
        nn.Linear(784, 250),
        nn.Tanh(),
        nn.Linear(250, 50),
        nn.Tanh(),
        nn.Linear(50, 10),
        nn.LogSoftmax()
    )

class ConvolutionNet(nn.Module):
    def __init__(self):
        super(ConvolutionNet, self).__init__()
        self.conv_layers = nn.Sequential(
            nn.Conv2d(channels, 8, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(8, 16, 3, padding=1)
        )
        self.linear_layers = nn.Sequential(nn.Linear(image_size//2*image_size//2 * 16, 10), 
                                           nn.LogSoftmax())
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.linear_layers(x)
        return x
```

<i> 2.1 Сравните графики обучения для полносвязной нейросети на методах Adam, NAG и SGD. </i>

```python
model = FullConnectedNetwork()
loss_fn = torch.nn.NLLLoss(size_average=True)


learning_rate = 1e-4

optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)


adam_loss_arr, adam_test_loss_arr = learn_on_model( images_train, labels_train, model, loss_fn, optimizer, images_test,\
                                                    labels_test,\
                                                    learning_rate,\
                                                    max_epoch_num=15, batch_size = 32)


model = FullConnectedNetwork()

loss_fn = torch.nn.NLLLoss(size_average=True)
learning_rate = 1e-4

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate) # momentum, there is no nag
sgd_loss_arr, sgd_test_loss_arr = learn_on_model( images_train, labels_train, model, loss_fn, optimizer, images_test,\
                                                    labels_test,\
                                                    learning_rate,\
                                                    max_epoch_num=15, batch_size = 32)


model = FullConnectedNetwork()

loss_fn = torch.nn.NLLLoss(size_average=True)
learning_rate = 1e-4

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9) # momentum, there is no nag
momentum_loss_arr, momentum_test_loss_arr = learn_on_model( images_train, labels_train, model, loss_fn, optimizer, images_test,\
                                                    labels_test,\
                                                    learning_rate,\
                                                    max_epoch_num=15, batch_size = 32)

```

```python

plt.plot(np.arange(len(adam_loss_arr)), adam_loss_arr, label = "adam", alpha= 0.3)
plt.plot(np.arange(len(sgd_loss_arr)), sgd_loss_arr, label = "sgd", alpha = 0.3)
plt.plot(np.arange(len(momentum_loss_arr)), momentum_loss_arr,
         label = "momentum", alpha = 0.3)

plt.legend(loc = "best")
plt.show()
```
